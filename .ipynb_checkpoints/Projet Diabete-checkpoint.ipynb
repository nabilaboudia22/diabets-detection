```python
# Séparer les variables explicatives et la variable cible
X = data.drop(columns=["Outcome"])
y = data["Outcome"]
```

```python
# augmentation du nombre d'observation du dataset 
# pour ameliorer les performances du modele 
# le nombre choisi est aleatoire vous pouvez mettre un autre chiffre mais il faut faire attention de ne pas tomber dans le sur apprentissage
# ça  reste des données synthetiques 
from imblearn.over_sampling import SMOTE
strategy = {0:4400, 1:4400}
oversample = SMOTE(sampling_strategy=strategy)
X, y = oversample.fit_resample(X, y)
```

```python
# Séparer les données en ensembles d'entrainement et de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Normaliser les données
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# creation du modele

model = Sequential()

# Ajouter les couches cachées avec des fonctions d'activation ReLU
model.add(Dense(units=128, activation='relu', input_dim=8)) #la couche d'entre input_dim=8 le reste c'est des couche caché et la derniere c la couche de sortie 
#model.add(Dropout(0.1))
model.add(Dense(units=64, activation='relu'))
#model.add(Dropout(0.1))
model.add(Dense(units=8, activation='relu'))
# Ajouter la couche de sortie avec une fonction d'activation sigmoïde
model.add(Dense(units=1, activation='sigmoid'))

# Compiler le modèle
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Former le modèle sur les données d'entraînement
history = model.fit(X_train, y_train, epochs=80, batch_size=50,validation_split=0.2 ) #fit c l'entrainement
```

```python
# Prédire les valeurs pour les données de test
y_pred = model.predict(X_test)
#on va tester le modele avec le modele de test
# Calculer la matrice de confusion
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred.round())
print(cm)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True,fmt="d",cmap='Blues')
```

```python
#calcule des mesure devaluation
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix,accuracy_score
score = model.evaluate(X_train, y_train)
print("Train loss:", score[0])
print("Train accuracy:", score[1])
Score_test = model.evaluate(X_test, y_test)
print("Test loss:", Score_test[0])
print("Test accuracy:", Score_test[1])
Accuracy_ANN = Score_test[1]
print("accuracy du modele ANN:", Accuracy_ANN)
Precision_ANN = precision_score(y_test, y_pred.round() , average="macro")
print("Precision : ", Precision_ANN)
Recall_ANN = recall_score(y_test, y_pred.round() , average="macro")
print("Recall:", Recall_ANN)
F1_ANN = f1_score(y_test, y_pred.round() , average="macro")
print("F1-Score:", F1_ANN)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred.round())) #2200 c l'ensemble de test
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
fpr_ann, tpr_ann, thresholds = roc_curve(y_test, y_pred)

# Tracer la courbe ROC
plt.plot(fpr_ann, tpr_ann)

# Ajouter une ligne diagonale représentant le modèle aléatoire
plt.plot([0, 1], [0, 1], 'k--')

# Ajouter un titre et des légendes aux axes
plt.title('Courbe ROC')
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')

# Calculer l'AUC-ROC et l'afficher sur le graphe auc roc taux d'accuracy
auc_ann = roc_auc_score(y_test, y_pred)
plt.text(0.7, 0.2, f'AUC-ROC = {auc_ann:.2f}')

# Afficher le graphe
plt.show()
```

```python
# une petite comparaison entre les deux modeles 
# en terme de Accuracy, Precision, F1-Score, Recall et la courbe ROC
# Intitialisation de la figure en deux axes 
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Comparaison des Modeles', fontsize=16, fontweight='bold')
fig.set_figheight(7)
fig.set_figwidth(14)
fig.set_facecolor('white')

# Première figure
## taille des bar
barWidth = 0.2
ANN_score = [Accuracy_ANN, Precision_ANN, Recall_ANN, F1_ANN]
RF_score = [accuracy_RF, precision_test_RF, Recall_test_RF, F1_test_RF]

## la position des bar sur l'axis X
r1 = np.arange(len(ANN_score))
r2 = [x + barWidth for x in r1]

## creation du plot
ax1.bar(r1, ANN_score, width=barWidth, edgecolor='white', label='ANN')
ax1.bar(r2, RF_score, width=barWidth, edgecolor='white', label='Random Forest')

## Configuration des axis X et y
ax1.set_xlabel('Mesures', fontweight='bold')
labels = ['Accuracy', 'Precision', 'Recall', 'F1']
ax1.set_xticks([r + (barWidth * 1.5) for r in range(len(ANN_score))], )
ax1.set_xticklabels(labels)
ax1.set_ylabel('Score', fontweight='bold')
ax1.set_ylim(0, 1)

## creation des legendes & titre
ax1.set_title('Performances du Modèle', fontsize=14, fontweight='bold')
ax1.legend()

# deuxieme figure
## Comparaison des courbes ROC des deux modeles
ax2.plot(fpr_ann, tpr_ann, label='ANN, auc = {:0.5f}'.format(auc_ann))
ax2.plot(fpr_RF, tpr_RF, label='Random Forest, auc = {:0.5f}'.format(auc_RF))

## Configuration des axis X et y 
ax2.set_xlabel('False Positive Rate', fontweight='bold')
ax2.set_ylabel('True Positive Rate', fontweight='bold')

## creation des legendes & titre
ax2.set_title('ROC Curve', fontsize=14, fontweight='bold')
ax2.legend(loc=4)

plt.show()

```
